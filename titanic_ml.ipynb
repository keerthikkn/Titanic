```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train.drop(columns=["Cabin"],inplace = True)
test.drop(columns=["Cabin"],inplace=True)
```

```python
train["Age"].replace(np.nan,train.Age.mean(),inplace=True)
test["Age"].replace(np.nan,test.Age.mean(),inplace=True)
```

```python
train.Embarked.mode()
```

```python
train["Embarked"].replace(np.nan,"S",inplace = True)
test["Fare"].replace(np.nan,train.Fare.mean(),inplace=True)
```

```python
target = pd.DataFrame(train,columns=["Survived"])
train.drop(columns=["Survived"],inplace=True)
```

```python
target.tail()
```

```python
train.drop(columns=["PassengerId","Name","Ticket"],inplace=True)
test.drop(columns=["PassengerId","Name","Ticket"],inplace=True)
```

```python
train=pd.get_dummies(data=train,columns=["Pclass","Sex","Parch","Embarked"],drop_first=True)
test=pd.get_dummies(data=test,columns=["Pclass","Sex","Parch","Embarked"],drop_first=True)
```

```python
test.head()
```

```python
test.drop(columns=["Parch_9"],inplace=True)
```

```python
train.head()
```

```python
train.describe()
```

```python
test.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
sc_train=pd.DataFrame(scale.fit_transform(train))
sc_test=pd.DataFrame(scale.transform(test))
```

```python
sc_test.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import PrecisionRecallDisplay, RocCurveDisplay
```

```python
X_train, X_test, y_train, y_test = train_test_split(sc_train,target,train_size=0.9,stratify=target)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, colorbar=False, cmap='Purples_r')
plt.grid(False)
```

```python
print(accuracy_score(y_test, y_pred),
precision_score(y_test, y_pred, average='macro'),
recall_score(y_test, y_pred, average='macro'),
f1_score(y_test, y_pred, average='macro'))
```

```python
survived = model.predict(sc_test)
```

```python
final = pd.DataFrame(test.PassengerId)
final["Survived"] = survived
```

```python
final.to_csv("prediction.csv",index = False)
```
